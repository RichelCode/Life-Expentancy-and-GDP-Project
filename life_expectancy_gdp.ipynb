# Further Univariate Analysis

### Addressing Skewness in Life Expectancy and GDP

Both **life expectancy** and **GDP** exhibit a **right-skewed distribution**, meaning most values are clustered at the lower end, with a few extremely high values pulling the distribution to the right.  

To improve interpretability and achieve a more **normalized distribution**, we apply a **log transformation**. This helps in:  
- **Reducing skewness** for better visualization.  
- **Enhancing comparability** across different values.  
- **Highlighting patterns** that may not be visible in the original scale.

We will now analyze the transformed distributions to observe any improvements in data representation.


```python
#Applying log transformations on GDP 
life_df["Log_GDP"] = np.log10(life_df["GDP"])

plt.figure(figsize=(10,5))
sns.histplot(life_df["Log_GDP"],kde=False,bins=20,color="blue")
plt.title("Distribution of Log-Transformed GDP")
plt.show()
```

### Log Transformation of GDP

The histogram above represents **log10-transformed GDP values**, helping to normalize the original right-skewed distribution. Key insights:

- **The transformation reduces skewness**, making GDP differences across countries easier to compare.
- **Extremely high GDP values are compressed**, revealing patterns among lower and mid-range GDP countries that were previously overshadowed.
- The presence of **gaps in the distribution** suggests economic groupings, possibly reflecting divisions between low-income, middle-income, and high-income nations.

This transformation ensures that GDP variations are more evenly distributed and suitable for further analysis.


```python
# Apply log transformation on the Life Expectancy
life_df = life_df[life_df["Life expectancy at birth (years)"] > 0]
life_df["Log_Life_Expectancy"] = np.log10(life_df["Life expectancy at birth (years)"])
```

```python
plt.figure(figsize=(10,5))
sns.histplot(life_df["Log_Life_Expectancy"],kde=False,bins=40,color="lightblue")
plt.title("Distribution of Log Life Expectancy")
```

### Life Expectancy Distribution Analysis

The histogram of **life expectancy** shows a distribution with distinct gaps, a **log transformation** was applied to reduce skewness, but it **did not significantly alter the distribution**. This suggests that:

- **Life expectancy values are already within a reasonable range**, making log transformation unnecessary.
- **The distribution is not as highly skewed as GDP**, so transformation does not impact it much.

We proceed with the **non-transformed life expectancy data** for further analysis.

This violin plot visualizes the distribution of **log-transformed GDP** across different countries, highlighting economic disparities and trends:

- The **white dot represents the median GDP**, while the **black bar shows the interquartile range (IQR)**.
- Countries with **longer violin shapes** have a **wider range of GDP values**.
- The plot reveals differences in GDP **spread and concentration** across nations.

- **United States**: The highest median GDP, with significant variability over time.
- **China**: Rapid economic growth with a wide distribution, showing increasing GDP.
- **Germany**: A stable economy with less variation in GDP.

```python
#Considering Life Expectancy across Countries
plt.figure(figsize=(14, 6))
sns.violinplot(data=life_df, x="Country", y="Life expectancy at birth (years)", palette="magma")

# Formatting
plt.xticks(rotation=90)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Life Expectancy (Years)", fontsize=12)
plt.title("Life Expectancy Distribution Across Countries", fontsize=14)
plt.show()

```

This violin plot compares the **distribution of life expectancy** across different countries:

```python
#Comparing Life Expectancy and Log GDP

plt.figure(figsize=(10,5))
sns.scatterplot(x="Life expectancy at birth (years)",y="Log_GDP",data=life_df,alpha=0.5)
plt.title("Relationship between Life Expectancy and GDP")
```

```python
# Compute correlation matrix
correlation_matrix = life_df[["Log_GDP", "Life expectancy at birth (years)"]].corr()
```

```python
# Set figure size
plt.figure(figsize=(6, 4))

# Create heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Add title
plt.title("Correlation Heatmap: GDP vs Life Expectancy")

# Show plot
plt.show()
```

### Correlation Between GDP and Life Expectancy

The heatmap above indicates a **strong positive correlation** between **GDP and Life Expectancy**, with a **correlation coefficient of 0.79**.  

This suggests that **countries with higher GDP tend to have longer life expectancy**, reinforcing the idea that economic growth contributes to better healthcare, improved living conditions, and overall well-being.  

While GDP is not the sole determinant of life expectancy, this correlation highlights the **economic disparities in health outcomes**, where wealthier nations generally have better access to medical advancements, nutrition, and infrastructure.
